```python
import glob
import pandas as pd

import re

from tqdm import tqdm
tqdm.pandas()

# pd.options.display.width = 1000
# pd.set_option('display.expand_frame_repr', False)
# pd.set_option('max_colwidth', 1000)
```

```python
df_list = []
chunks = glob.glob("collecting/data/pure_rwt/" + "*.csv")
assert len(chunks) > 0

for filename in chunks:
    df = pd.read_csv(
        filepath_or_buffer=filename,
        sep="⇧",
        engine="python",
        quotechar='"',
        on_bad_lines="warn",
        nrows=None
    )
    df_list.append(df)
df = pd.concat(df_list, axis=0)
df.reset_index(inplace=True, drop=True)
df
```

## Filter na columns

```python
df[df["column_data"].isna()]
```

```python
df = df[~df["column_data"].isna()]
df
```

```python
df = df[df["column_data"].str.strip() != ""]
df
```

## Remove `Unnamed: .`

```python
df.loc[:, "column_data"] = df["column_data"].progress_apply(
    lambda x: re.sub(r"Unnamed:\s\d+_level_\d+", " ", x)
)
df["column_data"].str.contains("Unnamed").sum()
```

```python
unnamed_mask = df["column_data"].str.contains("Unnamed")
df[unnamed_mask][:10]
```

## Remove: `.mw-parser`

```python
df.loc[:, "column_data"] = df["column_data"].progress_apply(
    lambda x: re.sub(r"@media.+{.+}", " ", x)
)
df["column_data"].str.contains("mw-parser").sum()
```

```python
# r"\.mw-parser-output \..+{.+}"
# r"\.mw-parser-output \.ts.+{.+}"

df.loc[:, "column_data"] = df["column_data"].progress_apply(
    lambda x: re.sub(r"\.mw-parser-output .+{.+}", " ", x)
)
df["column_data"].str.contains("mw-parser").sum()
```

```python
df[df["column_data"].str.contains("mw-parser")]
```

## Remove Wikipedia links

```python
df.loc[:, "column_data"] = df["column_data"].progress_apply(lambda x: re.sub(r"\[[0-9]*\]", "", x))
df
```

## Remove bad symbols

```python
"−" == "—"
```

```python
def remove_non_alphabetic_symbols(col_name: str) -> str:
    """TODO: just replace this symbols with space"""
    non_restricted_symbols = [
        "(",
        ")",
        "+",
        "-",
        "−",
        "[",
        "]",
        "/",
        ".",
        ",",
        "@",
        "{",
        "}",
        "«",
        "»",
        "<",
        ">",
        ":",
        ";",
        "'",
        "*",
        "?",
        "—",
        "!",
        "_",
    ]
    return "".join(c for c in col_name if c.isalpha() or c.isdigit() or c.isspace() or c in non_restricted_symbols)

df.loc[:, "column_data"] = df["column_data"].progress_apply(lambda x: remove_non_alphabetic_symbols(str(x)))
df
```

## Remove empty cells


```python
def remove_empty_cells(x: str) -> str:
    li = x.split(" << ")
    res = [i for i in li if not i.isspace() and i != ""]
    return " << ".join(res)

df.loc[:, "column_data"] = df["column_data"].progress_apply(remove_empty_cells)
df
```

## Remove columns that have len < 3 rows

```python
ge_3_rows_mask = df["column_data"].str.count("<<") >= 2
df[~ge_3_rows_mask]
```

```python
df = df[ge_3_rows_mask]
df
```

```python
def has_ge3_unique_cells(x: str) -> bool:
    li = x.split(" << ")
    unique_values = set(li)
    if len(set(li)) < 3:
        return False
    return True

has_ge3_unique_cells_mask = df["column_data"].progress_apply(has_ge3_unique_cells)
df[~has_ge3_unique_cells_mask]
```

```python
df = df[has_ge3_unique_cells_mask]
df
```
